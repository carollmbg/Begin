```python
#removi colunas com informações irrelevantes
#'Unnamed: 22': 'Unnamed: 23'= não tinham valores pertinentes
#'Case Number.2': 'Case Number.1:  'Case Number'= todos tem a informação da data do ocorrido e isso ja consta em date. E mesmo que fosse outro tipo de controle não seria interessante para essa analise.
#'pdf'= para essa analise não é necessario ter um documento externo para checagem
#'href formula': 'href'= endereço da fonte, como a base veio de uma fonte confiavel não achei necessario manter essa informação no caso a caso
shark2 = shark1.drop(columns=['Unnamed: 22',  'Unnamed: 23', 'Case Number.2', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number', 'original order' ])
```

```python
#da coluna age tirei tudo que não era numerico e com mais de 2 caracteres
#da coluna activity, tirei os textos duplicados e os substitui pela versão correta.
shark2['Age'].str.replace('.\D{2}.',"") 
shark2['Activity'].str.replace("Swimming","Swimming") 
shark2['Activity'].str.replace("Boogie boarding","Boogie Boarding")
shark2['Activity'].str.replace("Surf-skiing", "Surf skiing" )
```

```python
#tirei todos os valores não correspondentes a sexo da coluna
shark2.loc[(shark2['Sex '] !=  'M') & (shark2['Sex '] !=  'F') , 'Sex '] = ""
```

```python
#padronizei as especies white e tiger
shark2.loc[(shark2['Species '].str.contains('[Ww]hite')) & (shark2['Species '].notnull()), 'Species '] = "White Shark"
shark2.loc[(shark2['Species '].str.contains('[Tt]iger')) & (shark2['Species '].notnull()), 'Species '] = "Tiger Shark"
shark2.loc[(shark2['Species '] !=  "White Shark") & (shark2['Species '] != "Tiger Shark") , 'Species '] = "not defined"
```

```python
#separei todas as atividades que considerei relevantes
shark3= shark2.loc[shark2['Activity'].isin(['Swimming','Fishing', 'Spearfishing ', 'Bathing', 'Wading','Diving','Standing','Snorkeling','Scuba diving',
'Body boarding','Body surfing',  'Kayaking','Fell overboard','Pearl diving','Treading water','Free diving','Boogie Boarding', 'Windsurfing','Walking',  
'Shark fishing','Floating','Fishing','Canoeing', 'Surf-skiing', 'Surf-skiing', 'Surf fishing','Rowing','Surf skiing' ]) & shark2['Species '].isin(['White Shark', 'Tiger Shark', "not defined"])]
                                                                                                                                                    
```
